```
orderlines_df = pd.read_csv("data/raw/orderlines.csv")
```

```
orderlines_df["unit_price"] = orderlines_df["unit_price"].str.replace(".", "")
```

```
orderlines_df["unit_price"] = pd.to_numeric(orderlines_df["unit_price"])
```

```
combined_df = products_df.merge(orderlines_df, on="sku", how = "outer")
```

```
combined_df.info()
```

```
combined_df.groupby("sku")["unit_price"].nunique()
```

```
(
    combined_df[["sku", "name", "price", "promo_price", "unit_price"]]
 .groupby(["sku", "name", "price", "promo_price"])
 .agg(unit_price_max=("unit_price", "max"),
      unit_price_min=("unit_price", "min"))
 )
```

```
(
    combined_df[
        combined_df["price"].astype(str).str.contains(r"\.\d\d$", regex=True)
    ][["sku", "name", "price", "unit_price"]]
    .groupby(["sku", "name", "price"])
    .agg(
        unit_price_max=("unit_price", "max"),
        unit_price_min=("unit_price", "min")
    )
)
```

```
products_df["type"] = products_df["type"].str.replace(",", ".")
```

```
products_df["type"] = pd.to_numeric(products_df["type"])
```

```
products_df.info()
```

```
mask = products_df["price"].astype(str).str.contains(r"\.\d\d\d$", na=False)

products_df["price"] = products_df["price"].astype("object")

products_df.loc[mask, "price"] = (
    products_df.loc[mask, "price"]
    .astype(str)
    .str.replace(".", "", regex = False)
    .astype(float)
    / 10000
)
```
