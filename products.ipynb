```
import pandas as pd
```

```
products_df = pd.read_csv("data/raw/products.csv")
```

```
products_df
```

```
products_df = products_df.drop_duplicates()
```

```
products_df[products_df["sku"].duplicated(keep=False)]
```

```
products_df = products_df.drop(8000)
```

```
products_df.info()
```

```
products_df["type"] = products_df["type"].str.replace(",", ".")
```

```
products_df["type"] = pd.to_numeric(products_df["type"])
```

```
products_df.info()
```

```
mask = products_df["price"].astype(str).str.contains(r"\.\d\d\d\.\d\d\d$", na=False)

products_df["price"] = products_df["price"].astype("object")

products_df.loc[mask, "price"] = (
    products_df.loc[mask, "price"]
    .astype(str)
    .str.replace(".", "", regex = False)
    .astype(float)
    / 100000
)
```

```
products_df["price"] = pd.to_numeric(products_df["price"])
```

```
products_df
```

```
products_df = products_df.drop("promo_price", axis = 1)
```

```
products_df
```

```
products_df = products_df.dropna(subset = ["price"])
```

```
products_df.loc[products_df["desc"].isna()]
```

```
products_df.loc[products_df["desc"].isna(), "desc"] = \
    products_df.loc[products_df["desc"].isna(), "name"]
```

```
products_df["price"] = products_df["price"].round(2)
```

```
products_df.to_csv("data/processed/products.csv", index = False)
```
